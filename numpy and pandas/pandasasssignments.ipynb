
### Assignment 6: Time Series Analysis

1. Create a Pandas DataFrame with a datetime index and one column filled with random integers. Resample the DataFrame to compute the monthly mean of the values.
2. Create a Pandas DataFrame with a datetime index ranging from '2021-01-01' to '2021-12-31' and one column filled with random integers. Compute the rolling mean with a window of 7 days.


```python
start_date = '2022-01-01'
end_date = '2023-03-31'
date_range = pd.date_range(start=start_date, end=end_date, freq='D')
random_values = np.random.randint(10, 100, size=len(date_range))
df = pd.DataFrame({'Value': random_values}, index=date_range)
df_weakly_mean = df.resample('ME').mean()
df_weakly_mean
df.resample('ME').sum()
df.resample('ME').max()
```

```python
start_date = '2021-01-01'
end_date = '2021-12-31'
date_range = pd.date_range(start=start_date, end=end_date, freq='D')
random_values = np.random.randint(10, 100, size=len(date_range))
df = pd.DataFrame({'Temperature': random_values}, index=date_range)
df['Rolling_Mean_7D'] = df['Temperature'].rolling(window=7).mean()
df
```

### Assignment 8: Pivot Tables

1. Create a Pandas DataFrame with columns 'Date', 'Category', and 'Value'. Create a pivot table to compute the sum of 'Value' for each 'Category' by 'Date'.
2. Create a Pandas DataFrame with columns 'Year', 'Quarter', and 'Revenue'. Create a pivot table to compute the mean 'Revenue' for each 'Quarter' by 'Year'.


```python
num_entries = 20
start_date = '2023-01-01'
dates = pd.date_range(start=start_date, periods=num_entries, freq='D')
categories = ['A', 'B', 'C']

data = {
    'Date': np.random.choice(dates, size=num_entries),
    'Category': np.random.choice(categories, size=num_entries),
    'Value': np.random.randint(10, 100, size=num_entries)
}
df_category_value = pd.DataFrame(data)
df_category_value = df_category_value.sort_values(by='Date').reset_index(drop=True)
df_category_value

pivot_table_category_date_sum = df_category_value.pivot_table(
    values='Value',     
    index='Date',       
    columns='Category',  
    aggfunc='sum'       
)

pivot_table_category_date_sum
```

```python
num_entries = 30 
years = [2020, 2021, 2022]
quarters = [1, 2, 3, 4]

data = {
    'Year': np.random.choice(years, size=num_entries),
    'Quarter': np.random.choice(quarters, size=num_entries),
    'Revenue': np.random.randint(1000, 5000, size=num_entries)
}
df_revenue = pd.DataFrame(data)

df_revenue = df_revenue.sort_values(by=['Year', 'Quarter']).reset_index(drop=True)

pivot_table_revenue_mean = df_revenue.pivot_table(
    values='Revenue',     
    index='Year',         
    columns='Quarter',    
    aggfunc='mean'         
)
pivot_table_revenue_mean
```
